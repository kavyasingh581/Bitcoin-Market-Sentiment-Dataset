## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Convert 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])

# Convert 'timestamp' column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')

# Extract new features from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek # Monday=0, Sunday=6

# Display the first few rows with new features
display(df.head())
```

### What all manipulations have you done and insights you found?

I have performed the following data manipulations:

1.  **Converted Data Types**: The `date` column, initially an `object` type, has been converted to `datetime` objects using `pd.to_datetime()`. Similarly, the `timestamp` column (Unix timestamp) has also been converted to `datetime` objects for consistency and easier time-series analysis.
2.  **Extracted Time-Based Features**: From the `date` column, I extracted the following new features:
    *   `year`: The year of the record.
    *   `month`: The month of the record.
    *   `day`: The day of the month of the record.
    *   `day_of_week`: The day of the week (0 for Monday, 6 for Sunday).

These manipulations will allow for more granular analysis of trends over time, such as seasonal patterns or daily fluctuations in the Fear & Greed Index.

#### Chart - 5

```python
# Calculate the 7-day rolling average of 'value'
df['rolling_avg_7_day'] = df['value'].rolling(window=7).mean()

# Chart - 5: Line plot of 'value' and 'rolling_avg_7_day' over time
plt.figure(figsize=(15, 7))
sns.lineplot(x='date', y='value', data=df, label='Daily Value', color='lightcoral')
sns.lineplot(x='date', y='rolling_avg_7_day', data=df, label='7-Day Rolling Average', color='steelblue')
plt.title('Fear & Greed Index Value and 7-Day Rolling Average Over Time')
plt.xlabel('Date')
plt.ylabel('Fear & Greed Index Value')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

I chose a line plot to visualize the 'value' over time, combined with its rolling average. A line plot is ideal for time-series data to show trends. The rolling average smooths out short-term fluctuations, making the underlying trend clearer. Plotting both the raw data and the smoothed data allows for a direct comparison and better identification of sustained patterns versus temporary noise.

##### 2. What is/are the insight(s) found from the chart?

-   **Underlying Trends**: The rolling average line will reveal the smoother, longer-term trends in the Fear & Greed Index, making it easier to identify periods of sustained increase or decrease in sentiment, independent of daily volatility.
